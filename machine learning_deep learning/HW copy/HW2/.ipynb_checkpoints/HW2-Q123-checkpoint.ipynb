```python
######### Problem 2 #########
import numpy as np

svpath = 'HW2_dataset/HW2_Q2_dataset/Q2_data.npz'
data = np.load(svpath)

X = data['X']
Y = data['Y']

W = np.zeros((4,4))
B = np.zeros((4,1))
dLdW = {}

######### Number of Iterations #########
for p in range(1000):
    correct_num = 0    
    ########## Loop through 40 samples ##########
    for i in range(40):
        x = X[i]
        denominator = 0

        ## Accuracy Calculation ##
        highest_val = np.exp( np.dot( W[0],x ) + B[0] )
        highest_idx = 0
        
        ########## Loop through each class weight vector ##########
        for j in range(4):
            denominator += np.exp( np.dot( W[j],x ) + B[j] )
        for j in range(4):
            ## Initialize W and b gradients
            if j not in dLdW:
                dLdW[j] = {}
                dLdW[j]['val'] = np.zeros((1,4))
                dLdW[j]['b'] = 0

            ## softmax score
            numerator = np.exp( np.dot( W[j],x ) + B[j] )
            
            ## Accuracy Calculation ##
            if numerator > highest_val:
                highest_idx = j
                highest_val = numerator

            ## Gradient Calculations ##
            dLdW[j]['val'] = dLdW[j]['val'] + (numerator/denominator)*x
            dLdW[j]['b'] = dLdW[j]['b'] + (numerator/denominator)
            if Y[i] == j:
                dLdW[j]['val'] -= x
                dLdW[j]['b'] -= 1
        
        ## Accuracy rate ##
        if highest_idx == Y[i]:
            correct_num = correct_num + 1

    ## Back Propagation ##
    for i in range(4):
        dLdW[i]['val'] = dLdW[i]['val']/40
        dLdW[i]['b'] = dLdW[i]['b']/40
        W[i] = W[i] - 2*dLdW[i]['val']
        B[i] = B[i] - 2*dLdW[i]['b']
    if correct_num == 40 and p == 999:
        print("The Algorithm converges to a correct solution with step size of 2\
        and 1000 iterations.")
        print("The vector Weight:")
        print(W)
        print("The vector b:")
        print(B)
        break
```

```python
######### Problem 3-a #########
# The following Python code can be used to load the data for each data 
# set specified by the "dataind".

import numpy as np

svpath = 'HW2_dataset/HW2_Q3_dataset/Q3_data.npz'
data = np.load(svpath)

# X, array of shape [n_samples, n_features]
# Y, array of shape [n_samples]
X = data['X']
Y = data['Y']
trainX = data['X']
trainY = data['Y']
testY = np.zeros((40,))

W = np.zeros((4,2))
B = np.zeros((4,1))
dLdW = {}

correct_num = 0    
iter_n = 2500
######### Number of Iterations #########
for p in range(iter_n):
    correct_num = 0
    ########## Loop through 40 samples ##########
    for i in range(40):
        x = X[i]
        denominator = 0

        ## Accuracy Calculation ##
        highest_val = np.exp( np.dot( W[0],x ) + B[0] )
        highest_idx = 0
        
        ########## Loop through each class weight vector ##########
        for j in range(4):
            denominator += np.exp( np.dot( W[j],x ) + B[j] )
        for j in range(4):
            ## Initialize W and b gradients
            if j not in dLdW:
                dLdW[j] = {}
                dLdW[j]['val'] = np.zeros((1,2))
                dLdW[j]['b'] = 0

            ## softmax score ##
            numerator = np.exp( np.dot( W[j],x ) + B[j] )

            ## Accuracy Calculation ##
            if numerator > highest_val:
                highest_idx = j
                highest_val = numerator
            
            ## Gradient Calculations ##
            dLdW[j]['val'] += (numerator/denominator)*x
            dLdW[j]['b'] += (numerator/denominator)
            if Y[i] == j:
                dLdW[j]['val'] -= x
                dLdW[j]['b'] -= 1
        
        ## Accuracy rate ##
        if p == iter_n - 1:
            testY[i] = highest_idx
        if highest_idx == Y[i]:
            correct_num += 1

    ## Back Propagation ##
    for i in range(4):
        dLdW[i]['val'] = dLdW[i]['val']/40
        dLdW[i]['b'] = dLdW[i]['b']/40
        W[i] = W[i] - 0.5*dLdW[i]['val']
        B[i] = B[i] - 0.5*dLdW[i]['b']
print("After "+str(iter_n)+" iterations, we get "+str(correct_num)\
      + " correct classification.")

######## Visualization ########
[trainX_n,_]=trainX.shape
trainX=np.hstack((trainX,np.ones((trainX_n,1))))

Fig1=plt.figure(1)
plot_x=trainX[:,0:2]
plot_y=trainY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==0:
        plt.plot(plot_x[i,0],plot_x[i,1],'bo',markersize=15)
    elif plot_y[i]==1:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    elif plot_y[i]==2:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'ks',markersize=15)
plt.show()

Fig2=plt.figure(2)
plot_x=trainX[:,0:2]
plot_y=testY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==0:
        plt.plot(plot_x[i,0],plot_x[i,1],'bo',markersize=15)
    elif plot_y[i]==1:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    elif plot_y[i]==2:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'ks',markersize=15)
plt.show()

plt.figure(3)
plot_x=trainX[:,0:2]
plot_y=testY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==trainY[i]:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
plt.show()

print("It classifies decently, usually gives 82.5%-87.5% Accuracy\
Most of the time it converges to 85% Accuracy")
```

```python
######### Problem 3-b #########
# The following Python code can be used to load the data for each data 
# set specified by the "dataind".

import numpy as np

svpath = 'HW2_dataset/HW2_Q3_dataset/Q3_data.npz'
data = np.load(svpath)

# X, array of shape [n_samples, n_features]
# Y, array of shape [n_samples]
X = data['X']
Y = data['Y']
trainX = data['X']
trainY = data['Y']
testY = np.zeros((40,))

W = np.random.randn(4,2)
W2 = np.random.randn(4,1)
B = np.zeros((4,1))
B2 = np.zeros((4,1))

dLdW = {}

def sigmoid(W_j, x, B_j):
    f1 = np.dot( W_j,x ) + B_j
    sig = 1/( 1+np.exp(-f1) )
    return sig

## Initialization
correct_num = 0    
iter_n = 1000
step = 1
step2 = 1
mark = 0
m_list = []

######### Number of Iterations #########
for p in range(iter_n):
    correct_num = 0
    ########## Loop through 40 samples ##########
    for i in range(40):
        x = X[i]
        denominator = 0

        ## Accuracy Calculation ##
        sig = sigmoid(W[0], x, B[0])
        highest_val = np.exp( np.dot( W2[0],sig ) + B2[0] )
        highest_idx = 0
        
        ########## Loop through each class weight vector ##########
        for j in range(4):
            ## Calculate softmax denominator (sum)
            sig = sigmoid(W[j], x, B[j])
            denominator += np.exp( np.dot( W2[j],sig ) + B2[j] )
        for j in range(4):
            ## Initialize W and b gradients
            if j not in dLdW:
                dLdW[j] = {}
                dLdW[j]['val2'] = np.zeros((1,1))
                dLdW[j]['b2'] = 0
                dLdW[j]['val'] = np.zeros((1,2))
                dLdW[j]['b'] = 0

            ## softmax score ##
            sig = sigmoid(W[j], x, B[j])
            numerator = np.exp( np.dot( W2[j],sig ) + B2[j] )

            ## Accuracy Calculation ##
            if numerator > highest_val:
                highest_idx = j
                highest_val = numerator

            ## Gradient Calculations ##
            softmax = numerator/denominator
            dLdW[j]['val'] += softmax*W2[j]*(1-sig)*sig*x
            dLdW[j]['b'] += softmax*W2[j]*(1-sig)*sig
            dLdW[j]['val2'] += softmax*sig
            dLdW[j]['b2'] += softmax
            if Y[i] == j:
                dLdW[j]['val'] -= W2[j]*(1-sig)*sig*x
                dLdW[j]['b'] -= W2[j]*(1-sig)*sig
                dLdW[j]['val2'] -= sig
                dLdW[j]['b2'] -= 1
        
        ## Accuracy rate ##
        if p == iter_n - 1:
            testY[i] = highest_idx
        if highest_idx == Y[i]:
                correct_num += 1

    m_list.append(correct_num)
    ## Back Propagation ##
    for i in range(4):
        dLdW[i]['val'] = dLdW[i]['val']/40
        dLdW[i]['b'] = dLdW[i]['b']/40
        dLdW[i]['val2'] = dLdW[i]['val2']/40
        dLdW[i]['b2'] = dLdW[i]['b2']/40
        W[i] = W[i].T - step*dLdW[i]['val']
        B[i] -= step*dLdW[i]['b']
        W2[i] = W2[i].T - step2*dLdW[i]['val2']
        B2[i] -= step2*dLdW[i]['b2']
print("After "+str(iter_n)+" iterations, we get "+str(correct_num)\
      + " correct classification.")

######## Visualization ########
[trainX_n,_]=trainX.shape
trainX=np.hstack((trainX,np.ones((trainX_n,1))))

plt.figure(1)
plt.plot(range(iter_n), m_list)
plt.show()

plt.figure(2)
plot_x=trainX[:,0:2]
plot_y=trainY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==0:
        plt.plot(plot_x[i,0],plot_x[i,1],'bo',markersize=15)
    elif plot_y[i]==1:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    elif plot_y[i]==2:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'ks',markersize=15)
plt.show()

plt.figure(3)
plot_x=trainX[:,0:2]
plot_y=testY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==0:
        plt.plot(plot_x[i,0],plot_x[i,1],'bo',markersize=15)
    elif plot_y[i]==1:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    elif plot_y[i]==2:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'ks',markersize=15)
plt.show()

plt.figure(4)
plot_x=trainX[:,0:2]
plot_y=testY[:]
[n_plot,_]=plot_x.shape

for i in range(n_plot):
    if plot_y[i]==trainY[i]:
        plt.plot(plot_x[i,0],plot_x[i,1],'gx',markersize=15)
    else:
        plt.plot(plot_x[i,0],plot_x[i,1],'r*',markersize=15)
plt.show()

print("The performance is slightly better, usually gives us 90% Accuracy\
But since weights are randomized, very rarely it goes to 70%")
```
