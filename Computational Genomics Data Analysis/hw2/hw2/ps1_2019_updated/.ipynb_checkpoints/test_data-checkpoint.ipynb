```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression

from statsmodels.discrete.discrete_model import Logit
from statsmodels.stats.multitest import multipletests
from sklearn.linear_model import Ridge 
```

```python
# exercise 5

train_exp = pd.read_csv('exercise5/train_expression.csv', sep=',', index_col = 0)
train_y   = pd.read_csv('exercise5/phen_train.csv', sep=',', index_col = 0)

train_exp = np.array(train_exp[train_y.index])
train_y   = np.array(train_y['Basal_or_Luminal'])

test_exp = pd.read_csv('exercise5/test_expression.csv', sep=',', index_col = 0)
test_y   = pd.read_csv('exercise5/phen_test.csv', sep=',', index_col = 0)

test_exp = np.array(test_exp[test_y.index])
test_y   = np.array(test_y['Basal_or_Luminal'])




model_lg = LogisticRegression()
model_lg.fit(train_exp.transpose(), train_y.transpose())
yhat = model_lg.predict_proba(test_exp.transpose())
yhat = np.array([(yi[1] >= 0.5) * 1 for yi in yhat])

precision = np.sum((yhat == 1) & (test_y == 1)) / np.sum(yhat == 1)
recall    = np.sum((yhat == 1) & (test_y == 1)) / np.sum(test_y == 1)
print(precision, recall)
```

```python
# exercise 5 - first 10 genes

train_exp = pd.read_csv('exercise5/train_expression.csv', sep=',', index_col = 0)
train_y   = pd.read_csv('exercise5/phen_train.csv', sep=',', index_col = 0)

train_exp = np.array(train_exp[train_y.index])[:10, :]
train_y   = np.array(train_y['Basal_or_Luminal'])

test_exp = pd.read_csv('exercise5/test_expression.csv', sep=',', index_col = 0)
test_y   = pd.read_csv('exercise5/phen_test.csv', sep=',', index_col = 0)

test_exp = np.array(test_exp[test_y.index])[:10, :]
test_y   = np.array(test_y['Basal_or_Luminal'])




model_lg = LogisticRegression()
model_lg.fit(train_exp.transpose(), train_y.transpose())
yhat = model_lg.predict_proba(test_exp.transpose())
yhat = np.array([(yi[1] >= 0.5) * 1 for yi in yhat])

precision = np.sum((yhat == 1) & (test_y == 1)) / np.sum(yhat == 1)
recall    = np.sum((yhat == 1) & (test_y == 1)) / np.sum(test_y == 1)
print(precision, recall)
```

```python
## exercise 6 - MAF
genotypes = pd.read_csv('exercise6/genotype.csv', sep=',', index_col = 0)
genotypes = genotypes.iloc[:, :500]
AF = genotypes.apply(lambda x: np.sum(x) / len(x) / 2, axis=0)
MAF = AF.apply(lambda x: np.min([x, 1-x]))

print(np.sum(MAF > 0.03))
print(np.sum(MAF > 0.05))
print(np.sum(MAF > 0.1))
```

```python
# exercist 6 - logistic regression
phen = pd.read_csv('exercise6/phenotype.csv', sep=',', index_col = 0)
phen = np.array(phen['V1'])
genotypes = np.array(genotypes)
```

```python
genotypes = genotypes[:, np.where(MAF > 0.1)[0]]
genotypes.shape
```

```python
np.where(multipletests(np.array(pv), method = 'fdr_bh')[1] < 0.05)[0]
```

```python
# exercist 7 - ridge regression

train_exp = pd.read_csv('exercise7/train_expression.csv', sep=',', index_col = 0)
train_y   = pd.read_csv('exercise7/train_phen.csv', sep=',', index_col = 0)

train_exp = np.array(train_exp[train_y.index])
train_y   = np.array(train_y['V1'])

test_exp = pd.read_csv('exercise7/test_expression.csv', sep=',', index_col = 0)
test_y   = pd.read_csv('exercise7/test_phen.csv', sep=',', index_col = 0)
#test_exp = test_exp[test_exp.columns[:50]]

test_exp = np.array(test_exp[test_y.index])
test_y   = np.array(test_y['V1'])

```

```python
R2 = []
for alpha in [0.1, 1, 10, 100]:
    ml = Ridge(alpha=alpha)
    ml.fit(train_exp.transpose(), train_y)
    yhat = ml.predict(test_exp.transpose())
    R2.append(np.sum([x**2 for x in test_y - yhat]))
```

```python
np.argmin(np.array(R2))
```
